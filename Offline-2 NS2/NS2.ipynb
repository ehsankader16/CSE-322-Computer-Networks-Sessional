```python
import pandas as pd
import matplotlib.pyplot as plt
```

# **Varying Area**

```python
df = pd.read_csv('/data_varying_area.csv')
print(df) 
```

```python
plt.plot(df["Area"], df["Network throughput"], color ='m',marker='o')
plt.xlabel('Area(m^2)')
plt.ylabel('Network Throughput(bits/s)')
plt.xticks()
plt.grid(visible='true')
plt.show()
```

```python
plt.plot(df["Area"], df["End-to-end delay"], color ='b',marker='o')
plt.xticks()
plt.xlabel('Area(m^2)')
plt.ylabel('End-to-end delay(s)')
plt.grid(visible='true')
plt.show()
```

```python
plt.plot(df["Area"], df["Packet delivery ratio"], color ='g',marker='o')
plt.xticks()
plt.xlabel('Area(m^2)')
plt.ylabel('Packet delivery ratio')
plt.grid(visible='true')
plt.show()
```

```python
plt.plot(df["Area"], df["Packet drop ratio"], color ='r',marker='o')
plt.xticks()
plt.xlabel('Area(m^2)')
plt.ylabel('Packet drop ratio')
plt.grid(visible='true')
plt.show()
```
